```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
import pandas as pd
import numpy as ny
```

```python
url = "https://www.flipkart.com/oneplus-10r-sierra-black-128-gb/product-reviews/itm10b9b43146ada?pid=MOBGHBNGX9ZK6HQV&lid=LSTMOBGHBNGX9ZK6HQVDM5NXB&marketplace=FLIPKART"
```

```python
request = requests.get(url)
request
```

```python
request.content
```

```python
soup = bs(request.text)
soup
```

```python
comments = soup.find_all('div', {'class' : "t-ZTKy"})
for comment in comments:
    print(comment.get_text() + "\n\n")
```

```python
for i in range(1,9):
    request = requests.get(f"https://www.flipkart.com/oneplus-10r-sierra-black-128-gb/product-reviews/itm10b9b43146ada?pid=MOBGHBNGX9ZK6HQV&lid=LSTMOBGHBNGX9ZK6HQVDM5NXB&marketplace=FLIPKART&page={i}")
    soup = bs(request.text)
    comments = soup.find_all('div', {'class' : "t-ZTKy"})
    for comment in comments:
        print(comment.get_text() + "\n\n")
```

```python
for i in range (1,9):
    page = requests.get(f"https://www.flipkart.com/oneplus-10r-sierra-black-128-gb/product-reviews/itm10b9b43146ada?pid=MOBGHBNGX9ZK6HQV&lid=LSTMOBGHBNGX9ZK6HQVDM5NXB&marketplace=FLIPKART&page={i}")
    soup = bs(page.text)
    ratings = soup.find_all('div', {'class': '_3LWZlK _1BLPMq'})
    for rating in ratings:
        print(rating.get_text() + "\n\n")
```
