```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Heart.csv")
df
```

```python
df.isnull().sum()
```

```python
df.fillna(method = 'ffill', inplace = True)
```

```python
df.isnull().sum()
```

```python
(df.values == "?").sum()
```

```python
df = df.replace("?", np.nan)
```

```python
sns.histplot(df['Age'])
```
